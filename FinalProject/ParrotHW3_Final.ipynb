# Here is the part we build an index of the significant words on the onShape glossary page

```python
def fetch_page(url):
  response = requests.get(url)
  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup
  else:
    return None
```

```python
def index_words(soup):
  index = {}
  words = re.findall(r'\w+', soup.get_text())
  for word in words:
    word = word.lower()
    if word in index:
      index[word] += 1
    else:
      index[word] = 1
  return index
```

```python
def remove_stop_words(index):
  stop_words = {'a', 'an', 'the', 'and', 'or','in', 'on', 'at'}
  for stop_word in stop_words:
    if stop_word in index:
      del index[stop_word]
  return index
```

```python
def apply_stemming(index):
  stemmer = PorterStemmer()
  stemmed_index = {}
  for word, count in index.items():
    stemmed_word = stemmer.stem(word)
    if stemmed_word in stemmed_index:
      stemmed_index[stemmed_word] += count
    else:
      stemmed_index[stemmed_word] = count
  return stemmed_index
```

```python
def search(query, index):
  stemmer = PorterStemmer()
  query_words = re.findall(r'\w+', query.lower())
  results = {}
  for word in query_words:
    word = stemmer.stem(word)
    if word in index:
      results[word] = index[word]
  return results
```

```python
def search_engine(url, query):
  soup = fetch_page(url)
  if soup is None:
    return None
  index = index_words(soup)
  index = remove_stop_words(index)
  index = apply_stemming(index)
  results = search(query, index)
  return results
```

```python
def find_index(url):
  soup = fetch_page(url)
  if soup is None:
    return None
  index = index_words(soup)
  index = remove_stop_words(index)
  index = apply_stemming(index)
  return index
```

```python
def find_top_n_elements(objects, n=10):
    if len(objects) <= n:
        return sorted(objects, key=lambda x: x['freq'], reverse=True)

    # Initialize an empty list to store the top n elements
    top_n = []

    # Iterate through the array of objects
    for obj in objects:
        if len(top_n) < n:
            top_n.append(obj)
            top_n.sort(key=lambda x: x['freq'], reverse=True)
        else:
            if obj['freq'] > top_n[-1]['freq']:  # Compare with the smallest frequency in top_n
                top_n[-1] = obj
                top_n.sort(key=lambda x: x['freq'], reverse=True)

    return top_n
```

```python
#find the index for each word in the url , also find the most 10 words that have most frequent
def add_index_to_firebase(url):
    firebase_url = 'https://homework2-12bd2-default-rtdb.asia-southeast1.firebasedatabase.app/'
    FBconn = firebase.FirebaseApplication(firebase_url,None)
    result = find_index(url)
    index=[]
    for word,sum in result.items():
      data={
          'freq':sum,
          'term':word
      }
      index.append(data)
    top_10=find_top_n_elements(index)
    result = FBconn.put('/', 'Top_10_word', top_10)
    print(result)
    result = FBconn.put('/', 'Index', index)
    print(result)

add_index_to_firebase('https://cad.onshape.com/help/Content/Glossary/glossary.htm?tocpath=_____19')
```

```python
#find the index for 10 words selected
results = []
url = 'https://cad.onshape.com/help/Content/Glossary/glossary.htm?tocpath=_____19'

# Execute the search engine function for each word and store the results
query = ['point', 'keyboard', 'context', 'plan', 'type', 'draw', 'opens', 'allow', 'sketch', 'part']

for word in query:
    result = search_engine(url, word)
    results.append({
        'term': word,
        'freq': result
    })

index = []

for entry in results:
    data = {
        'term': entry['term'],
        'freq': entry['freq']
    }
    index.append(data)

result = FBconn.put('/', 'Selected_10_word', index)
print(result)
```

```python
def create_chatbot_page():
    custom_css = """
    <style>
        body {
            font-family: Arial, sans-serif;
            background-color: #f0f0f0;
            padding: 20px;
        }
        .container {
            max-width: 800px;
            margin: 0 auto;
            background-color: white;
            border-radius: 10px;
            padding: 20px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1);
        }
        h1 {
            color: #333;
            text-align: center;
        }
        #chat-output {
            height: 100px;
            overflow-y: auto;

            padding: 10px;
            margin-bottom: 10px;
        }
        #user-input {
            width: 100%;
            padding: 5px;
            margin-bottom: 10px;
        }
        #send-button {
            width: 100%;
            padding: 10px;
            background-color: #4CAF50;
            color: white;
            border: none;
            cursor: pointer;
        }
    </style>
    """

    display(HTML(custom_css))

    title = widgets.HTML(value="<h1>Chatbot</h1>")
    chat_output = widgets.HTML(value='<div id="chat-output"></div>')
    user_input = widgets.Text(placeholder='Type your message here...', layout=widgets.Layout(width='100%'))
    send_button = widgets.Button(description='Send', button_style='success', layout=widgets.Layout(width='100%'))

    # Add instructions for possible questions
    instructions = widgets.HTML(value="""
    <p><strong>Possible questions you can ask:</strong></p>
    <ul>
        <li>To get the users names , ask "users"</li>
        <li>To get the tabs in the database , ask "tabs"</li>
        <li>To get the total hours , ask "total hours"</li>
        <li>To get the number of the documents , ask "how many documents"</li>
        <li>To get when was the latest update , ask "latest update"</li>
        <li>To get the night shift for the workers , ask "night shifts"</li>
        <li>To get the weekend shift for the workers , ask "weekend shifts"</li>
    </ul>
    """)

    def on_send_button_click(b):
        question = user_input.value
        user_input.value = ''  # Clear input after sending
        response = answer_question(question, data)
        chat_output.value += f'<p><strong>You:</strong> {question}</p>'
        chat_output.value += f'<p><strong>Chatbot:</strong> {response}</p>'

    send_button.on_click(on_send_button_click)

    display(widgets.VBox([title, instructions, chat_output, user_input, send_button]))

    back_button = widgets.Button(description='Back to Main')
    back_button.on_click(show_main_page)
    display(back_button)


def is_night_shift(time_str):
    try:
        hour = datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S").hour
        return 20 <= hour or hour < 6
    except ValueError:
        return False

def is_weekend(time_str):
    try:
        day = datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S").weekday()
        return day >= 5  # 5 and 6 are Saturday and Sunday
    except ValueError:
        return False

def calculate_total_hours(data):
    total_hours = {}
    for entry in data:
        user = entry.get('User')
        if user:
            total_hours[user] = total_hours.get(user, 0) + 1
    return total_hours

def count_special_shifts(data):
    special_shifts = {'night': {}, 'weekend': {}, 'holiday': {}}
    for entry in data:
        user = entry.get('User')
        time = entry.get('Time')
        if user and time:
            if is_night_shift(time):
                special_shifts['night'][user] = special_shifts['night'].get(user, 0) + 1
            if is_weekend(time):
                special_shifts['weekend'][user] = special_shifts['weekend'].get(user, 0) + 1
    return special_shifts

def answer_question(question, data):
    question = question.lower()

    if "how many" in question and "documents" in question:
        return f"There are {len(data)} documents in the database."

    elif "latest update" in question:
        if data:
            latest_time = max(entry['Time'] for entry in data if 'Time' in entry)
            return f"The latest update was on {latest_time}."
        else:
            return "There are no updates in the database."

    elif "users" in question:
        users = set(entry['User'] for entry in data if 'User' in entry)
        return f"The users in the database are: {', '.join(users)}."

    elif "tabs" in question:
        tabs = set(entry['Tab'] for entry in data if 'Tab' in entry)
        return f"The tabs in the database are: {', '.join(tabs)}."

    elif "total hours" in question:
        total_hours = calculate_total_hours(data)
        return "\n".join([f"{user}: {hours} hours" for user, hours in total_hours.items()])

    elif "night shifts" in question:
        special_shifts = count_special_shifts(data)
        return "\n".join([f"{user}: {count} night shifts" for user, count in special_shifts['night'].items()])

    elif "weekend shifts" in question:
        special_shifts = count_special_shifts(data)
        return "\n".join([f"{user}: {count} weekend shifts" for user, count in special_shifts['weekend'].items()])
    else:
        return "I'm sorry, I don't have enough information to answer that question."
```
